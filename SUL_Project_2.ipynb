```python
Q1 = df_airfare.quantile(0.25)
Q3 = df_airfare.quantile(0.75)
IQR = Q3 - Q1
print(IQR)
```

```python
df_airfare_final=df_airfare_final.drop('City1', axis=1)
df_airfare_final=df_airfare_final.drop('City2', axis=1)
```

```python
df_airfare_final = df_airfare_final.drop('mla', axis=1)
```

```python
df_airfare_final = df_airfare_final.drop('lpa', axis=1)
```

```python
X = df_airfare_final.drop('Average_Fare', axis=1)
```

```python
X
```

```python
y = df_airfare_final[['Average_Fare']]
y
```

```python
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn import linear_model
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=1)
```

```python
from sklearn.linear_model import LinearRegression
regression_model = LinearRegression()
regression_model.fit(X_train, y_train)
```

```python
for idx, column_name in enumerate(X_train.columns):   
    print("coefficient for {} is {}".format(column_name, regression_model.coef_[0][idx]))
```

```python
intercept = regression_model.intercept_[0]

print("The intercept for our model is {}".format(intercept))
```

```python
regression_model.score(X_test, y_test)
```
