```
pd_dataset.drop(['RowNumber','CustomerId','Surname'] ,axis=1, inplace=True)
```

```
pd_dataset.drop(['Geography','Gender'] ,axis=1, inplace=True)
```

```


from tensorflow import feature_column
from tensorflow.keras import layers
from sklearn.model_selection import train_test_split
```

```
X = pd_dataset.drop(['Exited'] ,axis=1)
```

```
Y = pd_dataset[['Exited']]
```

```
X_train, X_test, y_train, y_test  = train_test_split(X, Y, test_size=0.2)

```

```
X_test.shape
```

```
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
train_normal_x=scaler.fit_transform(X_train)
test_normal_x=scaler.fit_transform(X_test)
```

```
#Initialize Sequential Graph (model)
model = tf.keras.Sequential()

#model.add(tf.keras.layers.Dense(1, input_shape=(10,)))
#Add Dense layer for prediction - Keras declares weights and bias automatically
model.add(tf.keras.layers.Dense(16, input_shape=(8,), activation='relu'))
model.add(tf.keras.layers.Dense(32, activation='relu'))

model.add(tf.keras.layers.Dense(2, activation='softmax'))

```

```
#Compile the model - add Loss and Gradient Descent optimizer
model.compile(optimizer='sgd', loss='mse',metrics=['accuracy'])

#Normalize the data
#model.add(tf.keras.layers.BatchNormalization())
```

```
y_train
```

```
y_train = tf.keras.utils.to_categorical(y_train, num_classes=2)
```

```
y_train
```

### I have tried mutiple option to create a model & demonstrated the final one


```
model.fit(train_normal_x, y_train,
          epochs=100)
```

```
model.summary()
```

```
predict=model.predict(X_test)
predict
```

```
loss_and_acc = model.evaluate(train_normal_x, y_train)
print(loss_and_acc)
```

```
y_pred = model.predict(test_normal_x, verbose=0)
y_pred_classes = model.predict_classes(test_normal_x, verbose=0)
print(y_pred)
```
